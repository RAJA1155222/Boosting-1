Q1. What is boosting in machine learning?

Boosting is an ensemble technique that combines multiple weak learners sequentially to form a strong learner.

The idea is to focus on the mistakes made by previous models and improve them in subsequent models.

Weak learners are typically simple models like shallow decision trees (stumps).

Q2. Advantages and limitations of boosting

Advantages:

Reduces bias and often improves accuracy over single models.

Works well with heterogeneous and complex datasets.

Focuses on difficult-to-predict samples, improving overall performance.

Can handle classification and regression tasks.

Limitations:

Sensitive to noisy data and outliers тЖТ may overfit.

Computationally slower than bagging due to sequential training.

Harder to interpret compared to a single model.

Q3. How boosting works

Train a weak learner on the training data.

Evaluate the model and identify misclassified samples.

Increase the weight of misclassified samples so the next learner focuses on them.

Train the next weak learner on the weighted data.

Repeat steps 2тАУ4 for a predefined number of estimators.

Combine all weak learners weighted by their accuracy to form a strong learner.

Q4. Different types of boosting algorithms

AdaBoost (Adaptive Boosting) тАУ the most common and widely used.

Gradient Boosting (GBM) тАУ uses gradient descent to minimize a loss function.

XGBoost тАУ optimized and regularized version of Gradient Boosting.

LightGBM тАУ faster, gradient-based boosting with leaf-wise tree growth.

CatBoost тАУ gradient boosting optimized for categorical features.

Q5. Common parameters in boosting algorithms
Parameter	Description
n_estimators	Number of weak learners (trees) to train
learning_rate	Shrinks the contribution of each weak learner
max_depth	Maximum depth of each weak learner tree
subsample	Fraction of samples used for training each learner (Gradient Boosting)
loss	Loss function to optimize (e.g., linear, exponential)

Q10. Effect of increasing the number of estimators in AdaBoost

Increasing n_estimators:

Can improve accuracy initially.

Reduces bias because more weak learners are combined.

Too many estimators тЖТ risk of overfitting (especially with noisy data).

Combine with learning_rate to control contribution of each learner.

Rule of thumb: Use cross-validation to choose optimal number of estimators and learning rate.
